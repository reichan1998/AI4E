5. Cho dict tỉ lệ sử dụng phương tiện giao thông: transport = {'car': 0.2, 'bus': 0.5, 'bike': 0.2, 'motorbike': 0.1}. Vẽ pie chart cho số liệu trên.

```python
transport = {'car': 0.2, 'bus': 0.5, 'bike': 0.2, 'motorbike': 0.1}
vehicles = list(transport.keys())
frequency = list(transport.values())

plt.pie(frequency, labels = transport, startangle = 90, counterclock = False)
plt.legend()
plt.title('Ti le su dung ca phuong tien')
plt.show()
```

```python
import matplotlib.pyplot as plt

transport = {'car': 0.2, 'bus': 0.5, 'bike': 0.2, 'motorbike': 0.1}
plt.figure(figsize = (8, 8))
explode = (0, 0.1, 0, 0.1)
plt.title('Ti le cac loai phuong tien')
plt.pie(transport.values(), labels = transport.keys(), explode = explode, startangle = 90, autopct = '%.2f%%')
plt.legend()
plt.savefig('abc.png')
plt.show()
```

6. Crawl dữ liệu điểm thi THPT Quốc Gia năm 2022, vẽ histogram điểm thi môn toán.


```python
import requests
score = []
for x in range(27000001, 27000001 + 500):
  scraping_url = "https://dantri.com.vn/thpt/1/0/99/" +  str(x) + "/2022/0.2/search-gradle.htm"
  payload = {}
  headers = {}
  response = requests.request("GET", scraping_url, headers = headers, data = payload)
  info = response.json()['student']
  score.append(info['toan'])
```

```python
score_new = []
for x in score:
  if x!= None:
    score_new.append(float(x))

plt.figure(figsize = (15, 10))

xint = range(0, 11)
plt.hist(score_new, bins = 10)
plt.xticks(xint)
plt.title('Biểu đồ phổ điểm thi THPT môn toán - năm 2022')
plt.xlabel('Điểm')
plt.ylabel("Số lượng học sinh")
plt.show()
```
